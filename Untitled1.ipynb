```python
def reporte_calidad(df, nombre):
    reporte = pd.DataFrame({
        "columna": df.columns,
        "tipo": df.dtypes.values,
        "nulos": df.isnull().sum().values,
        "%_nulos": (df.isnull().sum().values / len(df) * 100).round(2),
        "valores_unicos": [df[c].nunique() for c in df.columns],
    })
    reporte["fuente"] = nombre
    return reporte

reportes = pd.concat([
    reporte_calidad(mundial_2018, "2018"),
    reporte_calidad(mundial_2022, "2022"),
    reporte_calidad(mundial_2026, "2026"),
    reporte_calidad(catalogo, "catalogo"),
], ignore_index=True)
display(reportes)
```

```python
def extraer_numero_goles(valor):
    """'5 goles' -> 5 | '3' -> 3 | 's/d' -> NaN | 2.0 -> 2"""
    if pd.isna(valor):
        return np.nan
    if isinstance(valor, (int, float)):
        return valor
    texto = str(valor).strip().lower()
    if texto in ["s/d", "n/a", "na", "-", "", "none", "null", "sin dato"]:
        return np.nan
    match = re.search(r"\d+", texto)
    return int(match.group()) if match else np.nan

def parsear_fecha_flexible(valor):
    """Prueba varios formatos: 2018-06-14, 15/06/2018, 06/15/2018, 15-06-18, 'Jun 16, 2018'"""
    if pd.isna(valor):
        return pd.NaT
    texto = str(valor).strip()
    formatos = ["%Y-%m-%d", "%d/%m/%Y", "%m/%d/%Y", "%d-%m-%y", "%b %d, %Y"]
    for fmt in formatos:
        try:
            return pd.to_datetime(texto, format=fmt)
        except (ValueError, TypeError):
            continue
    try:
        return pd.to_datetime(dateparser.parse(texto, dayfirst=True))
    except Exception:
        return pd.NaT

def extraer_grupo(valor):
    """'Grupo A' -> 'A' | 'group-b' -> 'B' | ' A ' -> 'A'"""
    if pd.isna(valor):
        return np.nan
    match = re.search(r"[A-Ha-h]", str(valor))
    return match.group().upper() if match else np.nan

MAPA_FASE = {
    "fase de grupos": "Fase de grupos", "group stage": "Fase de grupos",
    "grupos": "Fase de grupos", "1ra fase": "Fase de grupos",
    "gs": "Fase de grupos", "fase grupos": "Fase de grupos",
    "group": "Fase de grupos",
}
def normalizar_fase(valor):
    if pd.isna(valor):
        return np.nan
    clave = str(valor).strip().lower()
    return MAPA_FASE.get(clave, str(valor).strip().title())

def normalizar_booleano(valor):
    """'Sí'/'1'/'TRUE' -> True | 'No'/'0'/'FALSE' -> False"""
    if pd.isna(valor):
        return np.nan
    texto = str(valor).strip().lower()
    if texto in ["sí", "si", "1", "true", "yes"]:
        return True
    if texto in ["no", "0", "false"]:
        return False
    return np.nan

mapa_equipos = dict(zip(
    catalogo["variante"].str.strip().str.lower(),
    catalogo["nombre_canonico"]
))
def canonicalizar_equipo(nombre):
    if pd.isna(nombre):
        return np.nan
    clave = str(nombre).strip().lower()
    return mapa_equipos.get(clave, str(nombre).strip().title())

def reporte_duplicados(df, columnas_clave):
    dups = df[df.duplicated(subset=columnas_clave, keep=False)]
    print(f"Duplicados encontrados (clave={columnas_clave}): {len(dups)}")
    return dups
```

```python
COLUMNAS_ESTANDAR = [
    "torneo", "id_partido", "fase", "grupo", "jornada", "fecha",
    "equipo_local", "equipo_visitante", "goles_local", "goles_visitante",
    "anfitrion_local", "estadio", "ciudad", "fuente"
]

df2018 = mundial_2018.rename(columns={
    "ID Partido": "id_partido", "Año": "torneo", "Fase": "fase", "Grupo": "grupo",
    "Jornada": "jornada", "Fecha": "fecha", "Equipo Local": "equipo_local",
    "Equipo Visitante": "equipo_visitante", "Goles Local": "goles_local",
    "Goles Visitante": "goles_visitante", "Anfitrión Local": "anfitrion_local",
    "Estadio": "estadio", "Ciudad": "ciudad", "Fuente": "fuente",
})[COLUMNAS_ESTANDAR]

df2022 = mundial_2022.rename(columns={
    "match_id": "id_partido", "WorldCup": "torneo", "stage": "fase", "group_name": "grupo",
    "match_day": "jornada", "date": "fecha", "local": "equipo_local", "visitor": "equipo_visitante",
    "home_score": "goles_local", "away_score": "goles_visitante", "home_host": "anfitrion_local",
    "venue": "estadio", "city_name": "ciudad", "source_url": "fuente",
})[COLUMNAS_ESTANDAR]

df2026 = mundial_2026.rename(columns={
    "match": "id_partido", "wc": "torneo", "round": "fase", "grp": "grupo",
    "md": "jornada", "played_on": "fecha", "home": "equipo_local", "away": "equipo_visitante",
    "HG": "goles_local", "AG": "goles_visitante", "host_h": "anfitrion_local",
    "source": "fuente",
})
df2026["estadio"] = np.nan
df2026["ciudad"] = np.nan
df2026 = df2026[COLUMNAS_ESTANDAR]

base_unificada = pd.concat([df2018, df2022, df2026], ignore_index=True)
print("Shape unificada:", base_unificada.shape)
base_unificada.head()
```

```python
base_unificada["fase"] = base_unificada["fase"].apply(normalizar_fase)
base_unificada["grupo"] = base_unificada["grupo"].apply(extraer_grupo)
base_unificada["fecha"] = base_unificada["fecha"].apply(parsear_fecha_flexible)
base_unificada["goles_local"] = base_unificada["goles_local"].apply(extraer_numero_goles)
base_unificada["goles_visitante"] = base_unificada["goles_visitante"].apply(extraer_numero_goles)
base_unificada["anfitrion_local"] = base_unificada["anfitrion_local"].apply(normalizar_booleano)
base_unificada["equipo_local"] = base_unificada["equipo_local"].apply(canonicalizar_equipo)
base_unificada["equipo_visitante"] = base_unificada["equipo_visitante"].apply(canonicalizar_equipo)

print("Nulos por columna:")
print(base_unificada.isnull().sum())

print("\nDuplicados por id_partido:", base_unificada.duplicated(subset="id_partido").sum())

# Equipos que no se encontraron en el catálogo (revisar manualmente)
equipos_todos = pd.concat([mundial_2018["Equipo Local"], mundial_2018["Equipo Visitante"],
                            mundial_2022["local"], mundial_2022["visitor"],
                            mundial_2026["home"], mundial_2026["away"]]).dropna().unique()
sin_catalogar = [e for e in equipos_todos if str(e).strip().lower() not in mapa_equipos]
print("\nEquipos sin coincidencia exacta en catálogo:", sin_catalogar)

base_unificada.head(10)
```

```python
local = base_unificada.rename(columns={
    "equipo_local": "equipo", "goles_local": "goles_favor", "goles_visitante": "goles_contra"
})[["torneo", "id_partido", "fecha", "fase", "grupo", "equipo", "goles_favor", "goles_contra"]]

visitante = base_unificada.rename(columns={
    "equipo_visitante": "equipo", "goles_visitante": "goles_favor", "goles_local": "goles_contra"
})[["torneo", "id_partido", "fecha", "fase", "grupo", "equipo", "goles_favor", "goles_contra"]]

base_larga = pd.concat([local, visitante], ignore_index=True)
print("Shape base_larga:", base_larga.shape)
base_larga.head()
```

```python
CONFIG_TORNEOS = {
    2018: {"equipos": 32, "partidos_grupos": 48},
    2022: {"equipos": 32, "partidos_grupos": 48},
    2026: {"equipos": 48, "partidos_grupos": 72},
}

def calcular_metricas_por_torneo(df):
    resumen = df.groupby("torneo").agg(
        goles_totales=("goles_favor", "sum"),
        partidos=("id_partido", "nunique"),
        equipos=("equipo", "nunique"),
    ).reset_index()
    resumen["goles_por_partido"] = (resumen["goles_totales"] / resumen["partidos"]).round(2)
    resumen["goles_por_equipo"] = (resumen["goles_totales"] / resumen["equipos"]).round(2)
    return resumen

def calcular_metricas_por_equipo(df):
    resumen = df.groupby(["torneo", "equipo"]).agg(
        partidos_jugados=("id_partido", "nunique"),
        goles_favor=("goles_favor", "sum"),
        goles_contra=("goles_contra", "sum"),
    ).reset_index()
    resumen["diferencia_goles"] = resumen["goles_favor"] - resumen["goles_contra"]
    resumen["goles_favor_por_partido"] = (resumen["goles_favor"] / resumen["partidos_jugados"]).round(2)
    resumen["pct_goles_favor_sobre_totales"] = (
        resumen["goles_favor"] / resumen.groupby("torneo")["goles_favor"].transform("sum") * 100
    ).round(2)
    return resumen

metricas_torneo = calcular_metricas_por_torneo(base_larga)
metricas_equipo = calcular_metricas_por_equipo(base_larga)

display(metricas_torneo)
display(metricas_equipo.sort_values("goles_favor_por_partido", ascending=False).head(10))
```

```python
def detectar_correlacion_sospechosa(df, columna_objetivo, umbral=0.95):
    correlaciones = df.corr(numeric_only=True)[columna_objetivo].drop(columna_objetivo)
    sospechosas = correlaciones[correlaciones.abs() > umbral]
    if len(sospechosas) > 0:
        print("⚠️ Posible fuga de información en:")
        print(sospechosas)
    else:
        print(f"No se detectaron correlaciones sospechosas (> {umbral}).")
    return sospechosas

detectar_correlacion_sospechosa(base_clasificacion, "avanzo_grupos")
```
